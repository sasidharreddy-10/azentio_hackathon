```python
import pandas as pd

# Replace with your CSV file path
file_path = "Lending_Loan_Portfolio_1000_Raw.xlsx"

# Read the CSV file
df = pd.read_excel(file_path, sheet_name="Loan_Portfolio_Raw", engine="openpyxl")

# Show all columns in outputs
pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", 1000)
pd.set_option("display.width", 1000)

print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

print("\nFirst 5 rows:")
print(df.head())

print("\nDataFrame info:")
df.info()

print("\nData types:")
print(df.dtypes)

print("\nSummary statistics (all columns):")
print(df.describe(include="all").T)

print("\nMissing values per column:")
print(df.isnull().sum())

print("\nUnique values per column:")
print(df.nunique(dropna=False))
```

```python
import os
import glob
import json

output_dir = "outputs"
merged_records = []

for path in sorted(glob.glob(os.path.join(output_dir, "*.jsonl"))):
    with open(path, "r", encoding="utf-8") as f:
        try:
            data = json.load(f)
        except json.JSONDecodeError:
            f.seek(0)
            data = [json.loads(line) for line in f if line.strip()]

    if isinstance(data, dict):
        data = [data]

    for item in data:
        if not isinstance(item, dict):
            continue
        merged_records.append({
            "prompt": item.get("prompt"),
            "completion": item.get("completion"),
        })

final_path = os.path.join(output_dir, "final_training_file.json")
with open(final_path, "w", encoding="utf-8") as f:
    json.dump(merged_records, f, indent=2, ensure_ascii=False)

print(f"Merged {len(merged_records)} records into {final_path}")
```
